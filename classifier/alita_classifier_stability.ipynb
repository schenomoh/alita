```python
#Generate a csv-like DataFrame called "df" and a "df_pivoted" one where each column is pivoted into rows
def gen():
    for i in range(33,126):
        yield "code("+chr(i)+")"
transco = {'DEV':'toto', 'HOMOL':'titi', 'UAT': 'tata', 'PRE': 'tutu', 'PROD': 'tete'}
x = gen()
data = []

for runid in range(0,5):
    dt=datetime.datetime.now()
    for env in 'DEV', 'HOMOL', 'UAT', 'PRE', 'PROD':
        data.append((runid, env, 'env_00',  env))
        data.append((runid, env, 'transco_01',  transco[env]))
        for i in range(2,5):
            data.append((runid,env, 'power_0'+str(i), i**2))
        for i in range(10,13):
            data.append((runid,env,'code_'+str(i), next(x)))
        data.append((runid,env, 'now_13',dt))

        
df_pivoted = pd.DataFrame(data, columns=["run_id", "env", "field", "value"])
df = df_pivoted.pivot_table(index=['run_id', 'env'], columns = ['field'], values = ['value'], aggfunc ={'value': lambda x: x})['value']
```

```python
#Calculate field stability

known_fields={}
for k in list(df_pivoted.field.unique()): known_fields[k] ={}
    
for group_by in ( ['field'], ['env', 'field']):
    for tmp, data in df_pivoted.groupby( group_by    ):
        known_fields[  data['field'].values[0]  ][  '/'.join(group_by)  ] = 1/len(data['value'].unique() )

res=pd.DataFrame(known_fields).T

res['stability'] = res.apply(lambda x:
          'constant' if x['field'] > 0.6 else 
              'env_specific' if x['env/field'] > 0.6 else
                  'variable'
          ,axis = 1
         )
res
```
